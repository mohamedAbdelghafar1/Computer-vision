```python
import cv2

videoCapture = cv2.VideoCapture(r'images\MyInputVid.avi')
fps = videoCapture.get(cv2.CAP_PROP_FPS)
size = (int(videoCapture.get(cv2.CAP_PROP_FRAME_WIDTH)),
int(videoCapture.get(cv2.CAP_PROP_FRAME_HEIGHT)))

videoWriter = cv2.VideoWriter(
'MyOutputVid4.avi', cv2.VideoWriter_fourcc('I','4','2','0'), fps, size)


success, frame = videoCapture.read()
while success:
    grayFrame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)  # Convert to grayscale
    videoWriter.write(grayFrame)                         # Write grayscale frame
    success, frame = videoCapture.read()
    
videoCapture.release()
videoWriter.release()
cv2.destroyAllWindows()
```

```python
import cv2
import os

videoCapture = cv2.VideoCapture('images\MyInputVid.avi')

output_folder ="frames"

c = 0
success, frame = videoCapture.read()

while success:
    name = os.path.join(output_folder, str(c) + '.jpg')
    print('Creating ... ' + name)
    cv2.imwrite(name, frame)
    c += 1
    videoWriter.write(frame)
    success, frame = videoCapture.read()

print(f"Finished processing {c} frames")
```

```python
import cv2

# Load the input video
videoCapture = cv2.VideoCapture('images\\MyInputVid.avi')

# Get video properties
fps = videoCapture.get(cv2.CAP_PROP_FPS)
size = (int(videoCapture.get(cv2.CAP_PROP_FRAME_WIDTH)),
        int(videoCapture.get(cv2.CAP_PROP_FRAME_HEIGHT)))

# Create a VideoWriter object for grayscale output
# Note: isColor=False tells OpenCV it's a single-channel (grayscale) video
videoWriter = cv2.VideoWriter(
    'MyOutputGrayVid6.avi',
    cv2.VideoWriter_fourcc(*'MJPG'),
    fps,
    size,
    isColor=True  # Must be True if you're writing 3-channel frames
)

# Read and process frames
success, frame = videoCapture.read()
while success:
    
    grayFrame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    grayFrame = cv2.cvtColor(grayFrame, cv2.COLOR_GRAY2BGR)  # Convert back to 3-channel
    videoWriter.write(grayFrame)                         # Write grayscale frame
    success, frame = videoCapture.read()

# Release resources
videoCapture.release()
videoWriter.release()
cv2.destroyAllWindows()
```

```python
import cv2

# Load the image directly in grayscale
img = cv2.imread(r"images\Screenshot 2025-09-01 182613.png", cv2.IMREAD_GRAYSCALE)

# Check if the image was loaded successfully
if img is None:
    print("Error: Image not found or path is incorrect.")
    exit()

# Display the grayscale image
cv2.imshow("Grayscale Image", img)

# Apply binary thresholding
thresh_value = 127
max_value = 255
_, blackAndWhiteImage = cv2.threshold(img, thresh_value, max_value, cv2.THRESH_BINARY)

# Display the binary image
cv2.imshow("Black and White Image", blackAndWhiteImage)

# Wait for a key press and close all windows
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import os

# Define the image path
image_path = r"images\Screenshot 2025-09-01 182613.png"

# Check if the file exists
if not os.path.exists(image_path):
    print(f"Error: File not found at {image_path}")
    exit()

# Load the image in grayscale
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Validate image loading
if img is None:
    print("Error: Failed to load image. Check the file format or path.")
    exit()

# Display the grayscale image
cv2.imshow("Grayscale Image", img)

# Apply binary thresholding
_, bw_image = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

# Display the binary image
cv2.imshow("Black and White Image", bw_image)

# Wait for key press and close windows
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
img =cv2.imread(r"images\Screenshot 2025-09-01 182613.png")

B, G, R = cv2.split(img)
cv2.imshow("OriginalImage", img)

cv2.imshow("Blue", B)
cv2.imshow("Green", G)
cv2.imshow("Red", R)

m=cv2.merge((B, G, R))
cv2.imshow("Merged", m)

cv2.waitKey()
cv2.destroyAllWindows()
```
